```python
import spacy
from spacy.lang.ro import Romanian
```

```python
nlp = spacy.load("en_core_web_md")
# nlp = Romanian()
```

```python
doc = nlp("This is a sentence. This is another sentence.")
for sent in doc.sents:
    print(sent.text)
```

```python
with open('ro_dedup100.txt', 'r') as ro_txt:
    for line in ro_txt:
        doc = nlp(line)
        for sent in doc.sents:
            print(sent.text)
```

```python
from nltk.tokenize import sent_tokenize
with open('ro_dedup100.txt', 'r') as ro_txt:
    for line in ro_txt:
        doc = sent_tokenize(line)
        for sent in doc:
            with open('ro_dedup100_processed.txt', 'a') as the_file:
                the_file.write("%s\n" % sent)
```
